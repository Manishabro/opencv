```python
data = {'emotion': {'angry': 52.97316312789917,
                     'disgust': 1.2661327541962919e-08,
                     'fear': 1.2440040707588196,
                     'happy': 0.0006009470325807342,
                     'sad': 34.86915826797485,
                     'surprise': 0.0038716938433935866,
                     'neutral': 10.909201204776764},
         'dominant_emotion': 'angry',
         'region': {'x': 206,
                    'y': 71,
                    'w': 113,
                    'h': 113,
                    'left_eye': None,
                    'right_eye': None},
         'face_confidence': 0.95,
         'age': 26,
         'gender': {'Woman': 0.06287714932113886, 'Man': 99.93712306022644},
         'dominant_gender': 'Man',
         'race': {'asian': 0.27394420467317104,
                  'indian': 0.011339824413880706,
                  'black': 0.0006015252893121215,
                  'white': 95.55931687355042,
                  'middle eastern': 1.5067902393639088,
                  'latino hispanic': 2.648013085126877},
         'dominant_race': 'white'}
```

```python
data['dominant_emotion']
```

```python
faceCascade=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
cv2.CascadeClassifier
```

```python
gray=cv2.cvtColor(imgg,cv2.COLOR_BGR2GRAY)
faces=faceCascade.detectMultiScale(gray,1.1,4)
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(imgg,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(imgg,
           data['dominant_emotion'],
           (0,50),
           font,1,
           (0,255,0),
           2,
           cv2.LINE_4);
```
